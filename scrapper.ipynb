```python
#Armo la funcion completa para scrapear cada link
from bs4 import BeautifulSoup

#Aerolinea y numero de vuelo
def get_vuelo(soup):
    a = []
    contentTable  = soup.find_all('div', { "class" : "ZCLkpe"}) # Use dictionary to pass key : value pair
    for x in contentTable:
        #print(x.text)
        a.append([x.text])
    #print(a)
    return a

#Horarios y puertas
def get_horarios(soup):
    r = []
    d = soup.find_all('tr', class_ ="kWOKDd")
    for i in range(0, len(d), 2):
        #print(d[i:i+2][1].text.split())
        #print('------------')
        r.append(d[i:i+2][1].text.split())
    return r

#Inicio --- Llegada
def get_inicio_llegada(soup):
    il = []
    d = soup.find_all('th', class_ ="a48lTe")
    for i in range(0, len(d), 2):
        #print([d[i:i+2][0].text, d[i:i+2][1].text])
        #print('-------------------')
        il.append([d[i:i+2][0].text, d[i:i+2][1].text])
    #print(il)
    return il


def scrapeo_vuelo(path):
    f = open(path, encoding='utf-8')    # simplified for the example (no urllib)
    soup = BeautifulSoup(f)
    f.close()
    ###
    v = get_vuelo(soup)
    h = get_horarios(soup)
    il = get_inicio_llegada(soup)
    data = []
    ###
    if ( len(v) == len(h) == len(il) ):
        for i in range(0, len(v)):
            data.append(v[i]+h[i]+il[i])
    ###
    return data
```

## Script final

```python
#Armo la funcion completa para scrapear cada link
import os
from bs4 import BeautifulSoup
import pandas as pd

#Aerolinea y numero de vuelo
def get_vuelo(soup):
    a = []
    contentTable  = soup.find_all('div', { "class" : "ZCLkpe"}) # Use dictionary to pass key : value pair
    for x in contentTable:
        #print(x.text)
        a.append([x.text])
    #print(a)
    return a

#Horarios y puertas
def get_horarios(soup):
    r = []
    d = soup.find_all('tr', class_ ="kWOKDd")
    for i in range(0, len(d), 2):
        #print(d[i:i+2][1].text.split())
        #print('------------')
        r.append(d[i:i+2][1].text.split())
    return r

#Inicio --- Llegada
def get_inicio_llegada(soup):
    il = []
    d = soup.find_all('th', class_ ="a48lTe")
    for i in range(0, len(d), 2):
        #print([d[i:i+2][0].text, d[i:i+2][1].text])
        #print('-------------------')
        il.append([d[i:i+2][0].text, d[i:i+2][1].text])
    #print(il)
    return il


def scrapeo_vuelo(path):
    f = open(path, encoding='utf-8')    # simplified for the example (no urllib)
    soup = BeautifulSoup(f)
    f.close()
    ###
    v = get_vuelo(soup)
    h = get_horarios(soup)
    il = get_inicio_llegada(soup)
    data = []
    ###
    if ( len(v) == len(h) == len(il) ):
        for i in range(0, len(v)):
            data.append(v[i]+h[i]+il[i])
    ###
    return data

def scrapear_html_aviones(path_carpeta):
    entries = os.listdir(path_carpeta)
    
    html_vacios = 0
    html_buenos = 0
    data2 = []
    for i in entries:
        try:
            r = scrapeo_vuelo(path_carpeta + i)
            if r:
                data2 += r
                html_buenos += 1
            else:
                html_vacios += 1
        except:
            print("Fallo en url: {}".format(path_carpeta + i))

    print("Archviso html buenos: {}".format(html_buenos))
    print("--------------------------------------------")
    print("Archviso html malos: {}".format(html_vacios))
    
    r = pd.DataFrame(data2, columns=columnas)
    r.to_pickle(path_carpeta + 'output.pkl')
    return r
    
```
